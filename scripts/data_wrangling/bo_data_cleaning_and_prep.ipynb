```python
import pandas as pd
import numpy as np

# Load the crime data
rate_of_aggravated_assault = pd.read_csv('..resources/raw_data/Rate of Aggravated Assault Offenses Per 100000 by Population_02-01-2024.csv.csv', skiprows=1)
rate_of_burglarly = pd.read_csv('../resources/raw_data/Rate of Burglary Offenses Per 100000 by Population_02-01-2024.csv', skiprows=1)
rate_of_larceny = pd.read_csv('../resources/raw_data/Rate of Larceny Offenses Per 100000 by Population_02-01-2024.csv', skiprows=1)
rate_of_motor_vehicle_theft = pd.read_csv('../resources/raw_data/Rate of Motor Vehicle Theft Offenses Per 100000 by Population_02-01-2024.csv', skiprows=1)
rate_of_property_crime = pd.read_csv('../resources/raw_data/Rate of Property Crime Offenses Per 100000 by Population_02-01-2024.csv', skiprows=1)
rate_of_violent_crime = pd.read_csv('../resources/raw_data/Rate of Violent Crime Offenses Per 100000 by Population_02-01-2024.csv', skiprows=1)

```

```python
# Convert csv to dataframe

ag_assault_df = pd.DataFrame(rate_of_aggravated_assault)
burglary_df = pd.DataFrame(rate_of_burglarly)
larceny_df = pd.DataFrame(rate_of_larceny)
motor_vehicle_theft_df = pd.DataFrame(rate_of_motor_vehicle_theft)
property_crime_df = pd.DataFrame(rate_of_property_crime)
violent_crime_df = pd.DataFrame(rate_of_violent_crime)



#Print the first 5 rows of the dataframe
ag_assault_df.head()
```

```python
#Transpose the dataframes
ag_assault_df = ag_assault_df.transpose()
burglary_df = burglary_df.transpose()
larceny_df = larceny_df.transpose()
motor_vehicle_theft_df = motor_vehicle_theft_df.transpose()
property_crime_df = property_crime_df.transpose()
violent_crime_df = violent_crime_df.transpose()

ag_assault_df

# Reset the index
ag_assault_df.reset_index(inplace=True)
burglary_df.reset_index(inplace=True)
larceny_df.reset_index(inplace=True)
motor_vehicle_theft_df.reset_index(inplace=True)
property_crime_df.reset_index(inplace=True)
violent_crime_df.reset_index(inplace=True)

ag_assault_df

```

```python
date_range = pd.date_range(start='1/1/1999', end='1/1/2023', freq='Y').year

# Create a new row named year
ag_assault_df['Year'] = date_range
burglary_df['Year'] = date_range
larceny_df['Year'] = date_range
motor_vehicle_theft_df['Year'] = date_range
property_crime_df['Year'] = date_range
violent_crime_df['Year'] = date_range

# Create a new column named US Aggravated Assault Rate by 100k
ag_assault_df['US Aggravated Assault Rate by 100k'] = ag_assault_df["index"]
burglary_df['US Burglary Rate by 100k'] = burglary_df["index"]
larceny_df['US Larceny Rate by 100k'] = larceny_df["index"]
motor_vehicle_theft_df['US Motor Vehicle Theft Rate by 100k'] = motor_vehicle_theft_df["index"]
property_crime_df['US Property Crime Rate by 100k'] = property_crime_df["index"]
violent_crime_df['US Violent Crime Rate by 100k'] = violent_crime_df["index"]

ag_assault_df.head()
```

```python
# Drop the index column
ag_assault_df.drop(columns='index', inplace=True)
burglary_df.drop(columns='index', inplace=True)
larceny_df.drop(columns='index', inplace=True)
motor_vehicle_theft_df.drop(columns='index', inplace=True)
property_crime_df.drop(columns='index', inplace=True)
violent_crime_df.drop(columns='index', inplace=True)


# Drop row 0
ag_assault_df.drop([0], inplace=True)
burglary_df.drop([0], inplace=True)
larceny_df.drop([0], inplace=True)
motor_vehicle_theft_df.drop([0], inplace=True)
property_crime_df.drop([0], inplace=True)
violent_crime_df.drop([0], inplace=True)

ag_assault_df
```

```python
# Combine the dataframes into one merging on the Year column
crime_rates_df = pd.merge(ag_assault_df, burglary_df, on='Year')
crime_rates_df = pd.merge(crime_rates_df, larceny_df, on='Year')
crime_rates_df = pd.merge(crime_rates_df, motor_vehicle_theft_df, on='Year')
crime_rates_df = pd.merge(crime_rates_df, property_crime_df, on='Year')
crime_rates_df = pd.merge(crime_rates_df, violent_crime_df, on='Year')

crime_rates_df.head()
```

```python
# Convert the data to a csv file
crime_rates_df.to_csv('crime_rates.csv', index=False)
```
